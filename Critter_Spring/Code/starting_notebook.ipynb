```python
import obspy
from obspy import read
import matplotlib.pyplot as plt
```

```python
st = read(r"../Data/1001.dat")
```

```python
print(st)
```

```python
scale = 1
plt.figure(figsize=(10, (72+2*scale)/8))

for ch in range(72):
    data=st[ch].data
    plt.plot(st[ch].times(), 0.8*data/abs(data).max()+ch, color='black') # add a separation between the wiggles

plt.title("Wiggle Plot")
plt.xlabel("Travel time (s)")
plt.ylabel("Channel")
plt.show()
```
