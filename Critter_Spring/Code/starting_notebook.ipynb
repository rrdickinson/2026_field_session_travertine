```python
import obspy
from obspy import read
import matplotlib.pyplot as plt
```

```python
st = read(r"../Data/1001.dat")
```

```python
print(st)
```

```python
# create image with pyplot.
# y axis of plot should go to (72+2*scale)*separation divided by 8000

scale = 1
separation = 1000
plt.figure(figsize=(10, ((72+2*scale)*separation)/8000))

for ch in range(72):
    data=st[ch].data
    plt.plot(0.8*separation*data/abs(data).max()+ch*separation, color='black') # add a separation between the wiggles
plt.show()
```
